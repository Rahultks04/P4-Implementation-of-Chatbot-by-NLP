```python
# Create the vectorizer and classifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression  # Import the missing module

vectorizer = TfidfVectorizer()
clf = LogisticRegression(random_state=0, max_iter=10000)

# Preprocess the data
tags = []
patterns = []
for intent in intents:
    for pattern in intent['patterns']:
        tags.append(intent['tag'])
        patterns.append(pattern)

# Training the model
x = vectorizer.fit_transform(patterns)
y = tags
clf.fit(x, y)
```

```python
# Track the last response for each tag
last_responses = {}

def chatbot(input_text):
    global last_responses

    # Transform the user input and predict the tag
    input_text = vectorizer.transform([input_text])
    tag = clf.predict(input_text)[0]

    # Find the intent corresponding to the tag
    for intent in intents:
        if intent['tag'] == tag:
            responses = intent['responses']
            
            # If the tag has been responded to before, remove the last response from the options
            if tag in last_responses and last_responses[tag] in responses:
                responses = [resp for resp in responses if resp != last_responses[tag]]
            
            # Choose a new response randomly
            response = random.choice(responses)
            
            # Update the last response for this tag
            last_responses[tag] = response
            return response
```

```python
import streamlit as st
counter = 0

def main():
    global counter
    st.title("Chatbot")
    st.write("Welcome to the library chatbot. Please type a genre to get recommendations on books.")

    counter += 1
    user_input = st.text_input("You:", key=f"user_input_{counter}")

    if user_input:
        # Check for "bye" or "goodbye" before invoking the chatbot function
        if user_input.lower() in ['bye', 'goodbye','thank you','thanks']:
            st.write("Thank you for chatting with me. Have a great day!")
            st.stop()
        
        # Get response from the chatbot function
        response = chatbot(user_input)
        st.text_area("Chatbot:", value=response, height=100, max_chars=None, key=f"chatbot_response_{counter}")
```
